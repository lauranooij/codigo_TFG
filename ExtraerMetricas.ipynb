```python
def primerPico(d1,bicepsEMG, tricepsEMG):
    
    minima_indices, _ = find_peaks(d1, distance=1500)  
    # minima_indices guarda los indices, no los valores
    
    # Asegurarse de tener al menos diez mínimos
    if len(minima_indices) < 10:
        raise ValueError("No hay suficientes mínimos para extraer 10 oscilaciones.")
    
    # Extraer la señal entre el primer mínimo y el quinto mínimo
    start_index = minima_indices[2]
    oscillation_segment = d1[start_index:len(d1)]  # +1 para incluir el punto final+
    #plt.figure(figsize=[16,4])
    #plt.plot(d1[start_index:len(d1)],label= 'ROM1')
    #plt.plot(bicepsEMG[start_index:len(d1)],label= 'biceps1')
    #plt.plot(tricepsEMG[start_index:len(d1)],label= 'triceps1')
    #plt.legend()
    return oscillation_segment,bicepsEMG[start_index:len(d1)] , tricepsEMG[start_index:len(d1)],  minima_indices[0]
```

```python
def angularVelocity(angles):
    a=angles
    times=[]
    timeStep=0.05
    for j in range (1, len(angles)+1,1):
        times.append(round(timeStep*j,3))
    deltaAngles = []
    deltaTimes = []
    angularVelocities = []
    
    # Calcular diferencias de ángulos y tiempos
    for i in range(len(angles) - 1):
        deltaAngles.append(angles[i+1] - angles[i])
        deltaTimes.append(times[i+1] - times[i]) #En este caso siempre 0.5s 
    
    delta_t = 0.05

    angular_speeds = np.diff(angles) / delta_t

    angular_speeds = np.concatenate(([0], angular_speeds))

    angularSpd = angular_speeds

    upsam = np.interp(
    np.linspace(0.0, 1.0, len(a), endpoint=False),  # where to interpret
    np.linspace(0.0, 1.0, len(a), endpoint=False),  # known positions
    angularSpd,  # known data points
    )
    upsamTimes = np.interp(
    np.linspace(0.0, 1.0, len(a), endpoint=False),  # where to interpret
    np.linspace(0.0, 1.0, len(a), endpoint=False),  # known positions
    times,  # known data points
    )
    
    return upsam,upsamTimes
```

```python
def capaturePeaksROM(time, ROM, plot=False):
    #print("ROM length:", len(ROM))
    arriba, _ = find_peaks(ROM, distance=1000)
    valles=[]
    for n in ROM:
        valles.append(-n)
    abajo, _ = find_peaks(valles, distance=1000)
    if plot: 
        plt.figure(figsize=(16, 4))
        for peak in arriba:
            plt.axvline(time[peak], linestyle='--', color='g', label="Flexionado" if peak == arriba[0] else "")
        for peak in abajo:
            plt.axvline(time[peak], linestyle='--', color='b', label="Extendido" if peak == abajo[0] else "")
        plt.plot(time, ROM, label="ROM")
        plt.legend()
    #print(len(arriba), len(abajo))

    return arriba,abajo
```

```python
def baseLine(tricespInicio, bicespInicio, indexBaseLine):
    #Cogemos baseline
    triceps = tricespInicio[0:indexBaseLine] #desde 0 hasta que empieza el ROM, hasta donde es el pico 0 
    biceps = bicespInicio [0:indexBaseLine]

    #minimo en baseline
    minValTrc = min(triceps)
    minValBic = min(biceps)

    indexTrc = np.where(triceps == minValTrc)[0][0] #Por si hay varios, accede al primero
    #indexTrc = np.where(triceps == minValTrc)[0][0]

    indexBic = np.where(biceps == minValBic)[0][0]
    #indexTrc = np.where(triceps == minValTrc)[0] #Por si hay varios, accede al primero
    #indexBic = np.where(biceps == minValBic)[0]
    
    #ventana 30 derecha, 30 izq
    if((indexTrc-30)<0):
        firstIndex=0
    else:
        firstIndex=indexTrc-30
        
    if((indexBic-30)<0):
        secondIndex=0
    else:
        secondIndex=indexBic-30

    
    windowTrc = triceps[(firstIndex):(indexTrc+30)]
    windowBic = biceps[(secondIndex):(indexBic+30)]
    
    # regla común en bioseñales para considerar que una señal es activa
    thresholdTrc = stat.mean(windowTrc) + 3* stat.stdev(windowTrc)
    thresholdBic = stat.mean(windowBic) + 3* stat.stdev(windowBic)
    return thresholdTrc, thresholdBic

def correctBaseLine(tricepsEMGadnorm, bicepsEMGadnorm, thresholdTrc, thresholdBic):
    tricepsEMGadnorm = tricepsEMGadnorm - thresholdTrc
    bicepsEMGadnorm = bicepsEMGadnorm - thresholdBic
     # Reemplazar valores negativos por 0
    tricepsEMGadnorm[tricepsEMGadnorm < 0] = 0
    bicepsEMGadnorm[bicepsEMGadnorm < 0] = 0
    #plt.figure(figsize=[16,4])
    #plt.plot(tricepsEMGadnorm,label= 'triceps', color = 'g')
    #plt.plot(bicepsEMGadnorm,label= 'biceps',color = 'orange') 
    #plt.legend()
    return tricepsEMGadnorm, bicepsEMGadnorm
```

```python
def calcular_ratio_rms(bic,tric):
    b = np.array(bic)
    rmsb = np.sqrt(np.mean(b**2))
    t = np.array(tric)
    rmst = np.sqrt(np.mean(t**2))
    b2= rmsb/(rmsb+rmst)
    t2= rmst/(rmsb+rmst)
    
    return b2, t2

def calcular_rms(bic,tric):
    b = np.array(bic)
    rmsb = np.sqrt(np.mean(b**2))
    t = np.array(tric)
    rmst = np.sqrt(np.mean(t**2))
    
    return rmsb, rmst
```

```python
user_id ="sabela"
def borrar_filas_por_id(csv_file, user_id):
    try:
        df = pd.read_csv(csv_file)

        if "user_id" not in df.columns:
            print(f"El archivo {os.path.basename(csv_file)} no tiene una columna 'user_id'.")
            return

        if user_id not in df["user_id"].values:
            print(f"'{user_id}' no se encuentra en {os.path.basename(csv_file)}.")
            return

        df_filtrado = df[df["user_id"] != user_id]
        df_filtrado.to_csv(csv_file, index=False)
        print(f"Filas con user_id '{user_id}' eliminadas de {os.path.basename(csv_file)}.")

    except Exception as e:
        print(f"Error al procesar {csv_file}: {e}")


borrar_filas_por_id(csv_ratioB, user_id)
borrar_filas_por_id(csv_ratioT, user_id)

borrar_filas_por_id(csv_rmsB, user_id)
borrar_filas_por_id(csv_rmsT, user_id)

borrar_filas_por_id(csv_meanB, user_id)
borrar_filas_por_id(csv_meanT, user_id)

borrar_filas_por_id(csv_medianB, user_id)
borrar_filas_por_id(csv_medianT, user_id)
```
