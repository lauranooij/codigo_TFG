```python
def linearMixto(df, y_label):
    # Reorganizar el DataFrame a formato largo
    df_long = pd.melt(df, id_vars=['user_id'],
                      var_name='tiempo',
                      value_name='valor')

    # Asegurarse de que tiempo sea tratado como categórico
    orden_tiempo = ['antes', 'despues', 'despues5', 'despues10','despues15','despues20','despues25']
    
    df_long['tiempo'] = df_long['tiempo'].astype('category')
    
    # Ajustar el modelo lineal mixto
    modelo = mixedlm("valor ~ tiempo", df_long, groups=df_long["user_id"])
    resultado = modelo.fit()
    print(df_long[df_long['tiempo'] == 'antes']['valor'].mean())

    # Mostrar resumen con más decimales
    with pd.option_context('display.float_format', '{:.6f}'.format):
        print(resultado.summary())
 
    # Mostrar p-values detallados
    print("\nP-values detallados:")
    print(resultado.pvalues.to_string(float_format='%.6f'))
    #save_pvalues_as_image(resultado.pvalues, f"{filename}_pvalues.png")

    # Calcular medias y errores estándar por tiempo
    resumen_plot = df_long.groupby('tiempo')['valor'].agg(['mean', 'std']).reset_index()
    resumen_plot.columns = ['tiempo', 'media', 'error']

    # Imprimir media ± std por tiempo
    print("\nMedia ± SD por tiempo:")
    for _, row in resumen_plot.iterrows():
        print(f"{row['tiempo']}: {row['media']:.3f} ± {row['error']:.3f}")

    # Ordenar el eje X por nombre
    orden_tiempo = ['antes','despues', 'despues5', 'despues10','despues15','despues20','despues25']
    resumen_plot['tiempo'] = pd.Categorical(resumen_plot['tiempo'], categories=orden_tiempo, ordered=True)
    resumen_plot = resumen_plot.sort_values('tiempo')
 
    # Crear la gráfica
    plt.figure(figsize=(12, 6))
    sns.lineplot(data=resumen_plot, x='tiempo', y='media', marker='o', color='black')
    plt.errorbar(resumen_plot['tiempo'], resumen_plot['media'],
                 yerr=resumen_plot['error'], fmt='none', ecolor='black', capsize=5)
    plt.xlabel('Time')
    plt.ylabel(y_label)
    # Diccionario para traducir los nombres solo en la gráfica
    nombre_ingles = {
        'antes': 'before',
        'despues': 'after',
        'despues5': 'after5',
        'despues10': 'after10',
        'despues15': 'after15',
        'despues20': 'after20',
        'despues25': 'after25'
    }

    # Aplicar etiquetas traducidas al eje X
    plt.xticks(ticks=range(len(resumen_plot['tiempo'])),
           labels=[nombre_ingles[val] for val in resumen_plot['tiempo']])   
    plt.grid(True)
    plt.tight_layout()  
    #plt.savefig(f"{filename}_plot.png", dpi=300)
    plt.show()
    plt.close()
    
    return resumen_plot
```

```python
def graficar_lmem_sesiones(resumenMFV, resumenC, filename, y_label):
    # Añadir etiquetas de grupo
    resumenMFV['grupo'] = 'MFV'
    resumenC['grupo'] = 'C'

    # Combinar los resúmenes
    resumen_total = pd.concat([resumenMFV, resumenC], ignore_index=True)

    estilos = {
        'MFV': {'color': 'darkorange', 'marker': 'o'},
        'C':   {'color': 'grey',       'marker': 's'}
    }

    orden_tiempo = ['antes', 'despues', 'despues5', 'despues10',
                    'despues15', 'despues20', 'despues25']

    resumen_total['tiempo'] = pd.Categorical(resumen_total['tiempo'], categories=orden_tiempo, ordered=True)
    resumen_total = resumen_total.sort_values('tiempo')

    nombre_ingles = {
        'antes': 'before', 'despues': 'after', 'despues5': 'after5', 'despues10': 'after10',
        'despues15': 'after15', 'despues20': 'after20', 'despues25': 'after25'
    }

    plt.figure(figsize=(12, 6))
    nombre_grupo = {'MFV': 'Intervention', 'C': 'Control'}


    for grupo, estilo in estilos.items():
        sub = resumen_total[resumen_total['grupo'] == grupo]
        if sub.empty:
            continue
        plt.plot(sub['tiempo'], sub['media'], label = nombre_grupo[grupo],
                 color=estilo['color'], marker=estilo['marker'])
        plt.errorbar(sub['tiempo'], sub['media'], yerr=sub['error'],
                     fmt='none', capsize=5, color=estilo['color'])

    plt.xlabel("Time (min)")
    plt.ylabel(y_label)
    plt.xticks(ticks=range(len(orden_tiempo)), labels=[nombre_ingles[t] for t in orden_tiempo])
    plt.title(y_label.split('mean')[0])
    plt.grid(True)
    plt.legend(title="Group")
    plt.tight_layout()
    plt.savefig(rf"C:\Users\laura\OneDrive - Fundación Universitaria San Pablo CEU\4.2\TFG\Final\V-resultados\plot_Resumen\{filename}.png", dpi=300)
    plt.show()

    
```

```python
def boxplot_function(df_MFV, df_C, y_label, filename):
    # Convertir a formato largo
    df_MFV = pd.melt(df_MFV, id_vars=['user_id'], var_name='tiempo', value_name='valor')
    df_C = pd.melt(df_C, id_vars=['user_id'], var_name='tiempo', value_name='valor')

    # Añadir etiquetas de grupo
    df_MFV['grupo'] = 'Intervention'
    df_C['grupo'] = 'Control'

    # Combinar
    df_total = pd.concat([df_MFV, df_C], ignore_index=True)

    
    
    # Ordenar los tiempos
    orden_tiempo = ['antes', 'despues', 'despues5', 'despues10','despues15', 'despues20', 'despues25']
    df_total['tiempo'] = pd.Categorical(df_total['tiempo'], categories=orden_tiempo, ordered=True)
    nombre_ingles = {
            'antes': 'before','despues': 'after', 'despues5': 'after5', 'despues10': 'after10',
            'despues15': 'after15', 'despues20': 'after20', 'despues25': 'after25'
        }
     # Calcular medias por grupo y tiempo
    medias = df_total.groupby(['tiempo', 'grupo'])['valor'].mean().reset_index()

    

    # Crear el boxplot
    plt.figure(figsize=(14, 6))
    sns.boxplot(
        data=df_total,
        x='tiempo',
        y='valor',
        hue='grupo',
        palette={'Intervention': 'orange', 'Control': 'lightgrey'},
        width=0.6,
        fliersize=2
    )
    
    plt.xlabel("Time (min)", fontsize=18)
    plt.ylabel(y_label, fontsize=18)
    plt.xticks(ticks=range(len(orden_tiempo)), labels=[nombre_ingles[t] for t in orden_tiempo], fontsize=14)
    plt.title(f"{y_label} values over time", fontsize=20)
    plt.grid(True)
    
    # Añadir líneas de medias discontinuas
    for grupo, estilo_color in zip(['Intervention', 'Control'], ['darkred', '#505050']):
        medias_grupo = medias[medias['grupo'] == grupo].sort_values('tiempo')
        plt.plot(
            medias_grupo['tiempo'],
            medias_grupo['valor'],
            linestyle='--',
            marker='o',
            color=estilo_color,
            label=f'{grupo} mean'
        )
        handles, labels = plt.gca().get_legend_handles_labels()
    by_label = dict(zip(labels, handles))
    plt.legend(by_label.values(), by_label.keys(), title="Grupo")
    #plt.ylim(top=1.4)
    plt.tight_layout()
    plt.savefig(rf"C:\Users\laura\OneDrive - Fundación Universitaria San Pablo CEU\4.2\TFG\Final\V-resultados\Boxplots\{filename}.png", dpi=300)
    plt.show()
```
